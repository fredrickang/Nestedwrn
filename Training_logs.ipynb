```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
log_dir = 'logs_test_0/test_0_error.csv'
```

```python
logs = pd.read_csv(log_dir)
```

```python
ax = plt.gca()

logs.plot(kind='line',x='step',y='train_err1',ax=ax)
logs.plot(kind='line',x='step',y='train_err2', color='red', ax=ax)
logs.plot(kind='line',x='step',y='train_err3', color='green', ax=ax)

plt.show()
```

```python
ag = plt.gca()

logs.plot(kind='line',x='step',y='val_err1',ax=ag)
logs.plot(kind='line',x='step',y='val_err2', color='red', ax=ag)
logs.plot(kind='line',x='step',y='val_err3', color='green', ax=ag)

plt.show()
```
