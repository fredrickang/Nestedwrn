```python
import numpy as np
import pickle as pk
import os
```

```python
test_dir = ['test1/','test2/']
logs_list = ['3-125.pk','3-1.pk','1-125.pk','1-1.pk']
```

```python
logs = [[] ,[]]
for i in range(len(test_dir)):
    for j in range(len(logs_list)):
        fo = open(os.path.join(test_dir[i],logs_list[j]),'rb')
        logs[i].append(pk.load(fo))
        fo.close()
```
