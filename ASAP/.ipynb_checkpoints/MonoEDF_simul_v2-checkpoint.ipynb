```python
import numpy as np
import random
import pickle as pk
from util import * ## taskset gen
from gcd import * ## calc gcd / lcm
```

```python
class task:
    def __init__(self,period,execution,deadline,idx):
        self.p = period
        self.e = execution
        self.d = deadline
        self.t_idx = idx # give numbering to task.
### !!! needs to be total_task[idx] == idx == j_idx !!! ###
# ex. n번째 테스크의 t_idx는 n이어야 하고, 해당 테스크의 잡들 역시 n값을 가져야 함.

class job:
    def __init__(self,start,wcet,deadline,exec_left_v,exec_left_r, idx):
        self.st = float(start)
        self.wcet = float(wcet)
        self.dl = float(deadline)
        self.exec_left_v = float(exec_left_v) # scheduler view. (for slack clac)
        self.exec_left_r = float(exec_left_r) # job view. (for job exit)
        self.j_idx = idx # match the numbering with task.
```

```python
# pure EDF Uniprocessor Scheduling simulation
# time value is expressed as milliseconds.

#global arrays.
total_tasks = []
cur_queue = []
ready_queue = [] # empty array. will be transformed to 2-dimensional ready_queue array.

# initialize tasks first
def init_taskset(period, execution, deadline, idx):
    m = task(period,execution,deadline,idx)
    return m

# initialize queue of ready jobs.
def init_readyqueue(flag=0):
    interval = LCMmerge(total_tasks)
    for i in range (len(total_tasks)):
        temp  = []
        job_num = interval / total_tasks[i].p
        for j in range (job_num):
            if (flag == 0):##WCET execution
                temp_job =job(j * total_tasks[i].p,
                              total_tasks[i].e,
                              j * total_tasks[i].p + total_tasks[i].d,
                              float(total_tasks[i].e),
                              float(total_tasks[i].e),
                              total_tasks[i].t_idx)
            elif (flag == 1):##realisitic execution.
                rseed = random.randrange(50,100)
                real_exec = total_tasks[i].e * float(rseed)/100
                temp_job =job(j * total_tasks[i].p,
                              total_tasks[i].e,
                              j * total_tasks[i].p + total_tasks[i].d,
                              float(total_tasks[i].e),
                              float(max(1.0,int(real_exec))) ,
                              total_tasks[i].t_idx)
            temp.append(temp_job)
        ready_queue.append(temp)
        

def enqueue_tasks(t_cur):
    # iterate over total_task configuration.
    # if it is time to initiate new job, append it to cur_queue.
    for i in range (len(total_tasks)):
        if ((t_cur % total_tasks[i].p) == 0):
            target_idx = total_tasks[i].t_idx
            if (len(ready_queue[target_idx]) != 0):##until queue becomes empty
                temp = ready_queue[target_idx].pop(0)
                cur_queue.append(temp)
            else:
                print("ready_queue empty!")

def dequeue_tasks(t_cur):
    # iterate over task queue.
    # if it is time to finish a job, delete it from list.
    remove_target = []
    for i in range (len(cur_queue)):
        if ((cur_queue[i].exec_left_r) == 0):
            #temporarily copy to another list.
            remove_target.append(cur_queue[i])
    # by looking at target list, delete finished jobs.
    for j in range (len(remove_target)):
        cur_queue.remove(remove_target[j])
        
def exec_tasks(t_cur):
    # iterate over task queue.
    # decrease exec_left value of "Earliest Deadline" task.
    if (len(cur_queue) != 0):
        elst_dead = cur_queue[0].dl
        exec_idx = 0
        for i in range (len(cur_queue)):
            if (((cur_queue[i].exec_left_r) != 0) and (cur_queue[i].dl <= elst_dead)):
                exec_idx = i
        cur_queue[exec_idx].exec_left_v -= 1
        cur_queue[exec_idx].exec_left_r -= 1
```

```python
def slack_calc(t_cur):
    #calc total utilization.
    u_tot = 0
    p = 0
    for i in range (len(total_tasks)):
        u_tot += float(total_tasks[i].e) / float(total_tasks[i].d)
        
    # make list of currently released jobs + earliest ready jobs.
    temp_q = []
    cur_job_idx = []
    # jobs in cur_queue
    for i in range (len(cur_queue)):
        temp_q.append(cur_queue[i])
        cur_job_idx.append(cur_queue[i].j_idx)
        print('cur_queue job appended')
    # jobs not unreleased
    for i in range (len(total_tasks)):
        if (total_tasks[i].t_idx in cur_job_idx):
            pass
        else:
            temp_q.append(ready_queue[i][0])
            print('ready job appended')
    #sort in order of shortest deadline.
    temp_q = sorted(temp_q, key=lambda job: job.dl)
    
    #calculate q and u.
    for j in reversed(range(len(temp_q))):
        #calculate minimum exec time for each task_i, in reverse order.
        u_tot -= temp_q[j].wcet /(temp_q[j].dl - temp_q[j].st)
        dl_diff = temp_q[j].dl - temp_q[0].dl
        print('loop=',j)
        print('u_tot is', u_tot)
        print('dl_diff is', dl_diff)
        if (dl_diff == 0):
            q_i = temp_q[j].exec_left_v
            u_tot = 1
        else:
            q_i = max(0,(temp_q[j].exec_left_v - (1-u_tot)*dl_diff))
            u_tot = min(1.0,u_tot + (temp_q[j].exec_left_v - q_i)/dl_diff)
        #add q_i to p
        print(',q_i=',q_i)
        p += q_i
    #total slack would be (interval - p)
    slack = temp_q[0].dl - t_cur - p
    return slack
```

```python
#initialize Tasks.
Task_1 = init_taskset(5,3,5,0)
Task_2 = init_taskset(7,2,7,1)
total_tasks.append(Task_1)
total_tasks.append(Task_2)
```
