```python
def print_tasks(Q):
    for task in Q:
        print("Task name: ",task.name,"period: ",task.p, "execution: ",task.e)
        
def print_jobs(Q):
    for jobs in Q:
        print("Job name: ",jobs.task.name,"period: ",jobs.task.p, "execution: ",jobs.task.e)
```

```python
def align_tasks(readyQ,execQ):
    readyQ_list =[]
    execQ_list = []
    sorted_tasks = []
    
    for jobs in readyQ: 
        readyQ_list.append(jobs.task.name)
        
    for jobs in execQ:  
        execQ_list.append(jobs.task.name)
    
    for names in execQ_list:
        readyQ_list.remove(names)
    
    for names in readyQ_list:
        for jobs in readyQ:
            if jobs.task.name == names:
                sorted_tasks.append(jobs)
    
    sorted_tasks = sorted_tasks + execQ
    sorted_tasks.sort(key = lambda jobs: jobs.d, reverse= True)
    return sorted_tasks

def calc_sim_limit(sorted_tasks):
    dl_nn_idx = 0
    
    for i in range (len(sorted_tasks)):
        if (sorted_tasks[i].task.t == True):
            dl_nn_idx = i
    if dl_nn_idx == (len(sorted_tasks)-1):
        sim_limit = -1
    else:
        sim_limit = sorted_tasks[len(sorted_tasks)-1].d
    
    return sim_limit, sorted_tasks[dl_nn_idx].d
```

```python
def slack_timing(t_cur, taskset):
    abs_deadline = t_cur + 33.0
    moduler = 33
    for task in taskset:
        tmp = abs_deadline % task.p
        if tmp == 0:
            moduler = task.p
        elif tmp < moduler:
            moduler = tmp
    target_deadilne = abs_deadline - moduler
    return target_deadilne
```

```python
def interference(slack_call, execQ, taskset):
    interval = 33 - slack_call
    full_action = 0
    carry_in = 0
    for tasks in taskset:
        full_action += math.floor(interval/tasks.p)*tasks.e
    
    for jobs in execQ:
        carry_in += jobs.remain
        
    return full_action + carry_in
```
