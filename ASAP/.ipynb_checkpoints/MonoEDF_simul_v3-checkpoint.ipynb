```python
import numpy as np
import random
import pickle as pk
from util import *
from gcd import *
```

```python
UNIT_TIME = 1
MODE2 = 0
MODE3 = 0
```

```python
spec = [(5,3),(7,2)]
```

```python
class task():
    def __init__(self, period, execution, name, is_target=False):
        self.p = period
        self.e = execution
        self.t = is_target
        self.name = name
        self.u = float(self.e/self.p)
```

```python
class job():
    def __init__(self,release,task):
        self.r = release
        self.task = task
        self.d = self.r + self.task.p
        self.ae = round((random.randrange(5,10,1)/10)* self.task.e,1)
        #self.ae = self.task.e
        self.remain = self.ae
        self.mode = 1
    def execute(self):
        self.remain = self.remain- UNIT_TIME
    def __lt__(self, other):
         return self.d < other.d
    def switch_mode(self, mode):
        if mode == 2:
            self.ae = MODE2
            self.mode = 2
        if mode ==3:
            self.ae = MODE3
            self.mode =3
```

```python
def init_sys(spec):
    print("========Init System========")
    readyQ = []
    execQ = []
    
    taskset = create_taskset(spec)

    
    for tasks in taskset:
        enqueue_job(readyQ, 0,tasks)
    return readyQ, execQ, taskset
```

```python
def create_taskset(spec):
    taskset = []
    num_task = len(spec)
    for i in range(num_task):
        taskset.append(task(spec[i][0],spec[i][1],str(i)))
    return taskset
```

```python
def enqueue_job(Q, future_release, task):
    job_instance = job(future_release, task)
    Q.append(job_instance)
    return Q
```

```python
def release_job(t_cur, readyQ, execQ):
    new_readyQ = []
    for jobs in readyQ:
        if jobs.r == t_cur:
            execQ.append(jobs)
            new_readyQ = enqueue_job(new_readyQ, t_cur+jobs.task.p, jobs.task)
        else:
            new_readyQ.append(jobs)
    return new_readyQ, execQ
```

```python
def ASAP_release_job(t_cur, readyQ, execQ):
    new_readyQ = []
    for jobs in readyQ:
        if jobs.r == t_cur:
            NN_released = jobs.t
            execQ.append(jobs)
            new_readyQ = enqueue_job(new_readyQ, t_cur+jobs.task.p, jobs.task)
        else:
            new_readyQ.append(jobs)
    return new_readyQ, execQ, NN_released
    
```

```python
def terminate_job(execQ):
    new_execQ = []
    for jobs in execQ:
        if jobs.remain > 0:
            new_execQ.append(jobs)
    return new_execQ 
```

```python
def execute(execQ):
    execQ.sort()
    if len(execQ) == 0:
        pass
    else:
        execQ[0].execute()
    return execQ
```

```python
def slack_calc(t_cur, taskset, readyQ, execQ):
    u_tot = 0
    p = 0 
    
    for tasks in taskset:
        u_tot += tasks.u
    
    readyQ_list =[]
    execQ_list = []
    
    sorted_tasks = []
    
    for jobs in readyQ:
        readyQ_list.append(jobs.task.name)
    for jobs in execQ:
        execQ_list.append(jobs.task.name)
    
    for names in execQ_list:
        readyQ_list.remove(names)
    
    for names in readyQ_list:
        for jobs in readyQ:
            if jobs.task.name == names:
                sorted_tasks.append(jobs)
    
    sorted_tasks = sorted_tasks + execQ
    sorted_tasks.sort(key =lambda jobs: jobs.d, reverse= True)
    
    for jobs in sorted_tasks:
        u_tot -= jobs.task.u
        d_diff = jobs.d - sorted_tasks[-1].d

        if d_diff == 0:
            qi =jobs.remain
            u_tot = 1
        else:
            qi = max(0,(jobs.remain - (1-u_tot)*d_diff))
            u_tot = min(1.0,u_tot +(jobs.remain))
        p += qi
    
    slack = sorted_tasks[-1].d - t_cur - p
    
    return slack
```
