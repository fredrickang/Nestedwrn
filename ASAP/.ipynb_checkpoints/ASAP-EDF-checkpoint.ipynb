```python
def ASAP_EDF(taskset_spec):
    
    readyQ, execQ, taskset = init_sys(taskset_spec)
    LCM = LCMmerge(taskset)

    time = 0
    exit_flag = 0

    nn_logs = []
    while True:
        if len(nn_logs) >10000:
            print("========End System========")
            break

        print('Time: ', round(time,1))

        # sync release
        #1. ready q에 있는 잡들중에 release time이 t_cur이랑 동일하면 execQ로 release
        #   1-1. release된 task는 새로운 잡을 생성해서 release time을 t_cur+period로 설정후 readyQ에 넣기
        #2. execQ 에 있는 job중에 remaining ex 가 0 인 잡들 제거 
        #3. exec for unit time 

        #ASAP 
        # if NN job release => call slack_calc
        # if slack is big enough => switch mode of NN
        # for [t_cur, t_cur+slack]  enforce to execute NN (slack = slack - overhead)
        # [t_cur+slack, ...] normal edf 


        # normal EDF
        # 서순 체크 필요
        #execQ = execute(execQ) #3
        #execQ = terminate_job(execQ) #2
        #readyQ, execQ = release_job(time, readyQ, execQ) #1 & 1-1


        #ASAP-EDF
        execQ = execute(execQ)
        execQ = terminate_job(execQ)
        readyQ, execQ, NN = ASAP_release_job(time, readyQ, execQ)

        #Deadline miss check
        for jobs in execQ:
            if(time > jobs.d):
                print("EDF scheduling fucked up!!")
                exit_flag = 1
        if(exit_flag == 1): 
            break

        # if NN task release
        if NN:
            for jobs in execQ:
                if jobs.task.t == True:
                    NN_job = jobs
            slack = slack_calc(time, taskset, readyQ, execQ)
            print("time",time)
            print(slack)
            if slack > BUDGET_2:
                NN_job.switch_mode(3)
            elif slack > BUDGET_1:
                NN_job.switch_mode(2)

            # execute during slack 
            for _ in range(int(slack)):
                time += UNIT_TIME
                print('Time: ', round(time,1))
                NN_job.execute()

                # if NN finish it's work
                if NN_job.remain == 0:
                    break

            for jobs in execQ:
                if jobs.task.t == True:
                    jobs = NN_job
            execQ = terminate_job(execQ)

            nn_logs.append(NN_job.mode)
        # if not
        else:
            time += UNIT_TIME


        print(" ")
        print("[execQ list]")
        for jobs in execQ:
            print("Task:",jobs.task.name," Remain jobs",jobs.remain)
        print(" ")
        time = round(time,1)

        print(" ")
    return nn_logs
```

```python
logs = []
for i in range(len(set_of_taskset)):
    tmp = []
    for j in range(len(set_of_taskset[i])):
        taskset_spec = set_of_taskset[i][j]
        taskset_spec.insert(0,NN_task)
        tmp.append(ASAP_EDF(taskset_spec))
    logs.append(tmp)
```
